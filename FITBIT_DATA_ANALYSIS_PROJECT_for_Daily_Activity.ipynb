ANALYSIS OF ACTIVITIES LEADING TO CALORIES BURNING

```python
X=dfDaily.iloc[:,2:-1]
Y=dfDaily.iloc[:,-1]
```

```python
X.head()
```

```python
Y.head()
```

**To analyze which features contribute more significantly to burning calories**


Since there are a lot of columns in the dataset, Chi square test has been used to select 5 most important features significantly contributing to calorie burning.



```python
#Selecting important activities contributing to burning calories - using Chi square test
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
best_features=SelectKBest(chi2,k=5)
dfDaily_best_features=best_features.fit_transform(X,Y)

print(dfDaily_best_features)
```

```python
X.head()
```

Comparing the best features values given by chi-square test to the values in the feature set to know the selected features

From the above 2 results, we can see that TotalSteps, VeryActiveMinutes, FairlyActiveMinutes, LightlyActiveMinutes, SedentaryMinutes are the 5 best selected features according to the Chi-square test

```python
dfDaily_new=dfDaily[['TotalSteps','VeryActiveMinutes','FairlyActiveMinutes','LightlyActiveMinutes','SedentaryMinutes']]
dfDaily_new.head()
```

```python
Y.head()
```

```python
#concatenating dfDaily_new and Y
merged_data=pd.concat([dfDaily_new,Y],axis=1)
merged_data.head()
```

**Creating a correlation heatmap for the 5 features selected by the Chi-square test along with 'Calories'**

This is done to see positive or negative correlation of each feature with respect to calorie burning

```python
#Creating a correlation heatmap on this merged_data

sns.heatmap(merged_data.corr(),annot=True)
```

```python
#Selecting 2 most important features contributing to burning calories - using Chi square test

best_2_features=SelectKBest(chi2,k=2)
dfDaily_best_2_features=best_2_features.fit_transform(X,Y)

print(dfDaily_best_2_features)
```

Best 2 features - TotalSteps, SedentaryMinutes

**Plotting a correlation heatmap to see positive or negative correlation of the above 2 selected features with Calories**

```python
dfDaily_new1=dfDaily[['TotalSteps','SedentaryMinutes','Calories']]
sns.heatmap(dfDaily_new1.corr())
```

**As per the above heatmap it can be concluded that
'SedentaryMinutes' has is negatively correlated to 'Calories' which
means that is Sedentary Minutes increases, Calorie burning will decrease.
Also it can be concluded that 'TotalSteps' is positively correlated to 'Calories' which means that as Total Steps increases, Calorie burning will also increase**
